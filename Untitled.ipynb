### Loading the dataset

we load the CSV file into a pandas DataFrame. We specify encoding='ISO-8859-1' due to potential encoding issues in the file and low_memory=False to avoid data type conflicts in large datasets.

```python
# load the dataframe
df = pd.read_csv('AviationData.csv', encoding='ISO-8859-1', low_memory=False)
df
```

### data overview

After loading the data, we inspect it to understand its structure, size, and any missing values:

```python
# accessing dataframe
df
```

```python
# Count of missing values in each column
df.isna().sum()
```

### Cleaning the Data

Since the dataset contains missing values, we set a threshold to drop columns with a high percentage of missing data. We remove columns where more than 30% of values are missing:

```python
# drop column that are not neccesary
threshold = 0.3
missing_values = df.isna().sum()

Total_rows = len(df)
drop_cols = [i for i in df.columns if missing_values [i] / Total_rows >threshold]

df.drop(columns=drop_cols, inplace=True)
                
```

After cleaning we will at any change in the columns and rows

```python
# Access the shape of dataframe
df.shape
```

Lets check for missing data in each column

```python
# Count of missing values in each column
df.isna().sum()
```

We make a copy for our new dataframe and called it df2

```python
# copy the dataframe
df2 = df.copy()
```

```python
# Access the shape of dataframe
df2.shape
```

### Cleaning the Data

cleaning row in the dataframe with missing values

```python
threshold = 1000
cols_with_missing = df2.columns[df2.isna().sum()>threshold]
df2.dropna(subset= cols_with_missing, inplace= True)
df2
```

```python
# Count of missing values
df2.isna().sum()
```

### Handling Remaining Missing Values

After dropping columns, we handle the remaining missing values in the important columns by filling them with the most frequent values (mode):

```python
# Fill missing values in Make columns
df2['Make'].fillna(df2['Make'].mode()[0], inplace=True);
```

```python
# Fill missing values in Location columns
df2['Location'].fillna(df2['Location'].mode()[0], inplace=True);
```

```python
# Fill missing values in Injury.Severity columns
df2['Country'].fillna(df2['Country'].mode()[0], inplace=True);
```

```python
# Fill missing values in Injury.Severity columns
df2['Injury.Severity'].fillna(df2['Injury.Severity'].mode()[0], inplace=True)
```

```python
# Fill missing values in Model columns
df2['Model'].fillna(df2['Model'].mode()[0], inplace=True)
```

```python
# lets check for missing value in each columns
df2.isna().sum()
```

## visualazation

### 1. Number of Accidents per Year

The following bar plot shows the number of accidents that occurred each year.

```python
df2['Event.Date'] = pd.to_datetime(df['Event.Date'], errors='coerce')
df2['Year'] = df2['Event.Date'].dt.year

Years = df2['Year'].value_counts()
plt.figure(figsize=(10,8))
plt.bar(x= Years.index, height= Years.values)
plt.title('Number of accidents per year')
plt.xlabel('Year')
plt.ylabel('number of accident')

plt.savefig('accident.png', format='png', dpi=300)
plt.show()
```

A decline in certain years can be indicative of improvements in safety regulations, technological advancements, or other external factors like increased/decreased air traffic.
If a trend is visible (e.g., a steady decline over the years), it could suggest that aviation safety has improved.
Alternatively, sudden peaks in accident frequency might warrant a deeper investigation into specific events or patterns during those years.

### 2. Accidents by Weather Condition

This line plot displays the number of accidents caused by different weather conditions.

```python
Weather = df['Weather.Condition'].value_counts()
plt.plot(Weather)
plt.title("Number of accident cost by weather condition")
plt.xlabel('Weather.Condition')
plt.ylabel('Number of accident')
plt.savefig('accident.png', format='png', dpi=300)
plt.show()
```

Some weather conditions, such as clear or cloudy skies, may be more frequently associated with accidents simply because they occur more often.
If severe weather conditions like thunderstorms or fog are highly represented, it could highlight the dangers of flying in adverse weather.
This plot could also reveal gaps in weather forecasting or a need for better training and equipment for flying under adverse conditions.

### 3. Total Fatal Injuries per Accident

A pie chart representing the number of total fatal injuries per accident.

```python
Injury = df2['Total.Fatal.Injuries'].value_counts().head()
plt.figure(figsize=(8,6))
plt.pie(Injury.values, labels=Injury.index, autopct='%1.1f%%', startangle=90, 
        labeldistance=1.1, pctdistance=0.85)

plt.title('Number of Total Fatal Injuries per Accident')

plt.axis('equal')
plt.savefig('accident.png', format='png', dpi=300)
# Display the plot
plt.show()
```

If the majority of accidents result in low or zero fatal injuries, this could indicate that most accidents are not severe or that modern aircraft are becoming more survivable.
A significant proportion of accidents with high fatalities would highlight more catastrophic events, which may call for further analysis into the causes and prevention strategies.
This information is crucial for safety regulators focusing on reducing fatalities through better equipment or emergency procedures.

### 4. Accidents by Engine Type

This scatter plot visualizes the number of accidents categorized by engine type.

```python
Engine = df2['Engine.Type']. value_counts()
plt.scatter(x= Engine.index, y= Engine.values)
plt.title('number of accident per engine type')
plt.xlabel('Engine.Type')
plt.xticks(rotation = 90)
plt.ylabel('number of accident')
plt.savefig('accident.png', format='png', dpi=300)
plt.show
```

Certain engine types may show higher accident frequencies, which could suggest they are more prone to failure, or simply that they are more commonly used.
By identifying trends across engine types, manufacturers and airlines can prioritize safety improvements or reconsider the use of particular engine models.
Correlations between engine types and accident frequency can also guide maintenance practices and inspections.

### 6. Survival Analysis over Engine Count

This scatter plot represents survival analysis, counting injury severity based on the number of engines in the aircraft.

```python
#Group the DataFrame by 'Number.of.Engines' and count the Injury.Severity for each airplane with a given number of engines.
Survialdf = df2.groupby('Number.of.Engines')['Injury.Severity'].size().rename('Injury count').reset_index()
Survialdf

plt.figure(figsize=(13,8))

# Creating a scatter plot with 'Number.of.Engines' on the x-axis and 'Injury count' on the y-axis
sns.scatterplot(data=Survialdf,x='Number.of.Engines',y='Injury count', s=100)

# Adding a title to the plot and displaying it
plt.title('Survival analysis over Engine count')
plt.savefig('accident.png', format='png', dpi=300)
plt.show()
```

The number of engines may influence the severity of an accident. For example, twin-engine aircraft may be more likely to avoid catastrophic failure compared to single-engine planes, leading to lower injury counts.
Aircraft with more engines could offer redundancy in case of failure, improving chances of survival.
This analysis could be useful for regulatory bodies when setting standards for engine redundancy or for airlines when considering fleet compositions.
